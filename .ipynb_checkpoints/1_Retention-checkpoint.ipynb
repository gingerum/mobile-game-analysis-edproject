Проверим, что дата регистрации = дате первой авторизации.

Для этого оставим в auth_df самую раннюю запись для каждого юзера, добавим к ним данные о регистрации и сравним даты.

```python
(auth_df
    .sort_values(by='auth_ts')
    .drop_duplicates(subset='uid')
    .merge(reg_df, on='uid', how='left')
    .assign(reg_auth_diff = lambda row: row.auth_ts == row.reg_ts)
    .reg_auth_diff.value_counts()
)
```

Отлично, это значит, что мы смело можем считать все авторизации в нулевой день полной когортой => рассчитывать % от них, а не от количества регистраций.

Заранее выделим сэмпл, на котором будем тестировать наши подсчеты - 2020-й год регистрации, за исключением неполного дня 23.09.

```python
reg_2020 = reg_df.query('reg_ts >= "2020-01-01" and reg_ts < "2020-09-23"')
auth_2020 = auth_df.query('auth_ts >= "2020-01-01" and auth_ts < "2020-09-23"')
```

## Функция для Classic Retention

Т.к. большая часть мобильных игр подразумевает ежедневную активность игроков, напишем функцию для подсчета классического удержания (по календарным дням).

На вход подаём датафреймы с регистрациями и входами, названия колонок с идентификаторами юзера, датами регистрации и входа, а также период, по которому нужно объединить юзеров в когорты.

```python
def classic_retention(reg_data, 
                      auth_data, 
                      users,
                      reg_date,
                      auth_date,
                      cohorts_period):
    
    df = reg_data.merge(auth_data, on=users)  # Данные о датах входа для всех зарегистрированных юзеров
    
    df['cohort'] = df[reg_date].dt.to_period(cohorts_period)  # Когорты от дат регистрации по заданному периоду
    
    df['retention_day'] = (df[auth_date] - df[reg_date]).dt.days  # Дни после регистрации для каждого входа
    
    # Сводная таблица с количеством уникальных входов в каждый день после регистрации по каждой когорте
    retention_count = df.pivot_table(index='cohort', columns='retention_day', values='uid', aggfunc='nunique')
    
    # Сводная таблица с долей входов в каждый день после регистрации относительно нулевого дня для каждой когорты
    retention_rate = retention_count.divide(retention_count.iloc[:,0], axis=0)
    
    return retention_count, retention_rate
```

Посмотрим на наши данные, сгруппировав игроков в когорты по месяцу регистрации.

```python
retention_count, retention_rate = classic_retention(
                                          reg_data = reg_2020, 
                                          auth_data = auth_2020, 
                                          users = 'uid',
                                          reg_date = 'reg_ts',
                                          auth_date = 'auth_ts',
                                          cohorts_period = 'M')
```

```python
retention_rate
```

#### Функции визуализации Classic Retention

```python
def heatmap_2020():
    with sns.axes_style("white"):
        fig, ax = plt.subplots(1, 2, figsize=(15, 6), sharey=True, gridspec_kw={'width_ratios': [1, 30]})

        # Выделим только первые 30 дней, т.к. ключевыми метриками для нас будут D1, D7 и D28(30)
        # День регистрации (D0) исключим, т.к. 100%
        sns.heatmap(retention_rate.iloc[:,1:31], 
                    mask=retention_rate.iloc[:,1:31].isnull(), 
                    annot=True,
                    annot_kws={"fontsize":8},
                    fmt='.1%', 
                    cmap='RdYlGn',
                    ax=ax[1])
        ax[1].set_title('Monthly Cohorts: User Retention (classic) by calendar days', fontsize=16)
        ax[1].set(xlabel='Days after registration',
                  ylabel='')

        # Размер когорт
        cohort_size_df = pd.DataFrame(retention_count.iloc[:,0]).rename(columns={0: 'cohort size'})
        sns.heatmap(cohort_size_df, 
                    annot=True,
                    annot_kws={"fontsize":8},
                    cbar=False, 
                    fmt='g',
                    cmap='RdYlGn',
                    ax=ax[0])

        fig.tight_layout()
```

#### Визуализация и выводы

```python
heatmap_2020()
```

**Хорошие новости** - % удержания относительно стабилен на протяжении первого месяца жизни игрока, а также MoM. <br>
